```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn
```

```python
fg = plt.figure()
# add_subplot可以为图像添加一个子图窗口, 2,2,1表示窗口是一个2*2的窗口，现在是第1个
ax1 = fg.add_subplot(2,2,1)
ax2 = fg.add_subplot(2,2,2)
ax3 = fg.add_subplot(2,2,3)

# 直方图
ax1.hist(randn(100), bins = 20, color = 'k', alpha = 0.3)
# 散点图
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
# 折线图
ax3.plot(randn(50).cumsum(), 'k--')
```

```python
# 在jupyter中如果想再次查看图像，直接调用该图像就可以了
fg
```

```python
# subplots是创建Figure + add_subplot的快捷方式
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(randn(500), bins=50, color='k', alpha=0.5)
# subplots_adjust用于调整图像间的边距
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
# 折线图
# 随机游走
randomWalker = randn(30).cumsum()
fg = plt.figure()
# add_subplot可以为图像添加一个子图窗口, 2,2,1表示窗口是一个2*2的窗口，现在是第1个
ax1 = fg.add_subplot(2,2,1)
ax2 = fg.add_subplot(2,2,2)
ax3 = fg.add_subplot(2,2,3)

# 一般表示法
ax1.plot(randomWalker, 'ko--')
ax2.plot(randomWalker, 'k--')
ax3.plot(randomWalker, '--')
```

```python
# 带参数的表示方法 color 颜色, linestyle表示线的类型, marker表示点的标记, 如上面所述的'k--'其中k为颜色表示灰色, --表示序列
ax1.plot(randomWalker, linestyle='--', color = 'g', marker = 'o')
ax2.plot(randomWalker, linestyle='dashed', color = 'g', marker = 'o')
ax3.plot(randomWalker, '--')
fg
```

```python
# drawstyle可以表示填充类型, 其中steps-post表示梯度的方式填充
plt.plot(randomWalker, linestyle='--', color = 'g', marker = 'o', drawstyle='steps-post')
```

```python
# label可以为每一条曲线添加标注
plt.plot(randomWalker, linestyle='-', color = 'g', marker = 'o', label = 'default')
plt.plot(randomWalker, linestyle='--', color = 'r', marker = 'o', drawstyle='steps-post', label = 'steps-post')
# 要加入legend，否则上面的将不生效，legend是生成图例的函数
plt.legend(loc='best')
# xlim可以限定图标x轴的范围
plt.xlim([1, 10])
```

```python
randomWalker = randn(1000).cumsum()
fg = plt.figure()
ax = fg.add_subplot(1, 1, 1)
ax.plot(randomWalker, linestyle='--', color = 'g', marker = 'o', label = 'default')
# plt.set_xticks([0, 100, 500, 1000])无法实现，需要是subplot类型
# set_xticks可以在图中标记刻度位置
ax.set_xticks([0, 100, 500, 1000])
```

```python
# set_xticklabels会为输入的xticks依次添加标签, 标签数和xticks数一致，多退少补
ax.set_xticklabels(['one', 'tow', 'three'])
# 设置图的标题
ax.set_title("random walker")
fg
```
